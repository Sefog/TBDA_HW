## Exercise 19: 
Create a stacked bar plot representing the following data:
<br>
`{'Apples': {'bad':30, 'good':15,'best':40},`
<br> 
`'Bananas':{'bad':15, 'good':35,'best':20},`
<br>
`'Cherries':{'bad':55, 'good':30,'best':45}}`

```python
data = {
    'Apples': {'bad': 30, 'good': 15, 'best': 40},
    'Bananas': {'bad': 15, 'good': 35, 'best': 20},
    'Cherries': {'bad': 55, 'good': 30, 'best': 45}
}


fruits = list(data.keys())
categories = list(data[fruits[0]].keys())  


bad_values = [data[fruit]['bad'] for fruit in fruits]
good_values = [data[fruit]['good'] for fruit in fruits]
best_values = [data[fruit]['best'] for fruit in fruits]

bar_width = 0.05

plt.bar(fruits, bad_values, label='Bad')
plt.bar(fruits, good_values, bottom=bad_values, label='Good')
plt.bar(fruits, best_values, bottom=[bad + good for bad, good in zip(bad_values, good_values)], label='Best')

for fruit, bad, good, best in zip(fruits, bad_values, good_values, best_values):
    plt.text(fruit, bad / 2, str(bad), ha='center', va='center')
    plt.text(fruit, bad + good / 2, str(good), ha='center', va='center')
    plt.text(fruit, bad + good + best / 2, str(best), ha='center', va='center')

plt.legend()


plt.show()
```

## Exercise 22:
Add the boxplot
<br>`IQR = Q3 - Q1`
<br>`lower_whisker = data[data >= Q1 - 1.5 * IQR].min()`
<br>`upper_whisker = data[data <= Q3 + 1.5 * IQR].max()`

```python
Q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1
lower_whisker = data[data >= Q1 - 1.5 * IQR].min()
upper_whisker = data[data <= Q3 + 1.5 * IQR].max()

plt.boxplot(data)

plt.plot(1, Q1, 'ro')  
plt.text(1.1, Q1, 'Q1: {:.2f}'.format(Q1), fontsize=10, va='center')

plt.plot(1, median, 'go')  
plt.text(1.1, median, 'Median: {:.2f}'.format(median), fontsize=10, va='center')

plt.plot(1, Q3, 'bo') 
plt.text(1.1, Q3, 'Q3: {:.2f}'.format(Q3), fontsize=10, va='center')

plt.plot(1, lower_whisker, 'b_')
plt.text(1.1, lower_whisker, 'Lower Whisker: {:.2f}'.format(lower_whisker), fontsize=10, va='center')

plt.plot(1, upper_whisker, 'r_')
plt.text(1.1, upper_whisker, 'Upper Whisker: {:.2f}'.format(upper_whisker), fontsize=10, va='center')

plt.show()

```

## Exercise 23:
Add the boxplot
<br>`lower_bound = Q1 - 1.5 * IQR`
<br>`upper_bound = Q3 + 1.5 * IQR`
<br>`outliers = data[(data < lower_bound) | (data > upper_bound)]`






```python
Q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_whisker = data[data >= Q1 - 1.5 * IQR].min()
upper_whisker = data[data <= Q3 + 1.5 * IQR].max()
outliers = data[(data < lower_bound) | (data > upper_bound)]

plt.boxplot(data)

plt.plot(1, Q1, 'ro')  
plt.text(1.1, Q1, 'Q1: {:.2f}'.format(Q1), fontsize=10, va='center')

plt.plot(1, median, 'go')  
plt.text(1.1, median, 'Median: {:.2f}'.format(median), fontsize=10, va='center')

plt.plot(1, Q3, 'bo') 
plt.text(1.1, Q3, 'Q3: {:.2f}'.format(Q3), fontsize=10, va='center')

plt.plot(1, lower_whisker, 'b_')
plt.text(1.1, lower_whisker, 'Lower Whisker: {:.2f}'.format(lower_whisker), fontsize=10, va='center')

plt.plot(1, upper_whisker, 'r_')
plt.text(1.1, upper_whisker, 'Upper Whisker: {:.2f}'.format(upper_whisker), fontsize=10, va='center')

plt.plot(1, lower_bound, 'b_')
plt.text(1.1, lower_bound, 'Lower Bound: {:.2f}'.format(lower_bound), fontsize=10, va='center')

plt.plot(1, upper_bound, 'r_')
plt.text(1.1, upper_bound, 'Upper Bound: {:.2f}'.format(upper_bound), fontsize=10, va='center')

plt.scatter(np.ones_like(outliers), outliers, color='red', label='Outliers')

plt.show()

```

## Exercise 27:
Create a 3D line plot for the function
<br>`x = np.linspace(-6, 6, 100)`
<br>`y = np.linspace(-6, 6, 100)`
<br>`z = x^2 + y^2`

```python
def func(x, y):
    return x**2 + y**2


x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
x, y = np.meshgrid(x, y)


z = func(x, y)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')


plt.show()

```
